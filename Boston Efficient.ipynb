```python
import numpy as np

data=np.loadtxt('training_boston.csv',delimiter=',')
data.shape
```

```python
x=data[:,:13]
y=data[:,-1]
```

```python
import pandas as pd
df=pd.DataFrame(x)
df
```

```python
#adding dummy columns to make function complex and effcient

df1=df.copy()
l=len(df.columns)
l
for i in range(l):
    for j in range(i,l):
        a=df.columns[i]
        b=df.columns[j]
        df1['b'+str(a)+str(b)]=df[df.columns[i]]*df[df.columns[j]]
x=df1.to_numpy()
print(df1.shape)
df1
```

```python
# featuring data
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()  # making a scaler mean=0 variance=1
scaler.fit(x)  #  scaler learning from x data and storing
x=scaler.transform(x)   # putting learned scaler to scale data of 'x'
```

```python
#converting numpy to dataframe
df=pd.DataFrame(x)
df
```

```python
# Adding a feature for intercept 'c' with values 1
df_new=df.copy()
df_new['c']=1

# lets make x and y from the dataset df_new
x=df_new.values

#values of y will remain same
```

```python
# lets train the data using the gradient descent funcction for multi-variables
def step_gradient_bostan(x,y,learning_rate,m):
    n_feature=x.shape[1]
    m_slope=np.zeros(n_feature)  # initializing initial slope to 0
    M=len(x)
    for i in range(M):
        for j in range(x.shape[1]):
            m_slope[j]= m_slope[j] + (-1/M) * ( y[i] - ( m * x[i]).sum() )*x[i][j]
        
    m_new = m - learning_rate * m_slope
    
    return m_new
```

```python
# this function will calculate total cost or error for each iteration

def cost(x,y,m):
    tcost=0
    M=len(x)
    for i in range(M):
        tcost = tcost + (1/(2*M)) * (( y[i] - (sum(m*x[i]) ))**2  )
    return tcost
```

```python
# this function will calculate m and c for n number of iterations.

def gd_multi_bostan(x,y,learning_rate,number_of_iteration):
    n_feature=x.shape[1]
    m=np.zeros(n_feature)  # initilizing m for 1st iteration taking random value of m as 0
        
    for i in range(number_of_iteration):
        m=step_gradient_bostan(x,y,learning_rate,m)
        print(i,'Cost:',cost(x,y,m))
    return m
        
```

```python
# Lets make a run function which will run gradient descent function for n number of iterations and having learning rate
# returns m,c

def run_bostan(x,y):
    learning_rate=0.05
    
    number_of_iteration=500

# if the no of itration will be more than a limit then there will be overshooting and you will get nan values
    # as if youn keep num_of_itration is 1000 then  it will give nan for run
    m=gd_multi_bostan(x,y,learning_rate,number_of_iteration)
    return m
    
```

```python
m=run_bostan(x,y)
print(m)
```
